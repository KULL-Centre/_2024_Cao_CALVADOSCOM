```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.lines import Line2D
plt.rcParams.update({ 'font.size': 12 })
```

### OPN

PRE rates for recombinant quail OPN220 protein (Q9I832_COTJA): residues 46-264 + vector-encoded start methionine. Data kindly provided by Robert Konrat and Andreas Beier, University of Vienna (DOI: 10.1039/c5cp04858c).

```python
df_opn = pd.read_csv('pre_rates.csv',index_col=1).iloc[44:,1:]

columns = pd.MultiIndex.from_tuples([(int(i[0][1:])-44,'value') if len(i)==1 else (int(i[0][1:])-44,'error') for i in df_opn.columns.str.split('_')],
                                   names=['label','type'])

df_opn = pd.DataFrame(df_opn.values,columns=columns,index=df_opn.index-44)
```

```python
labels = df_opn.columns.levels[0]
labels
```

```python
!mkdir expPREs
for label,df in df_opn.groupby(level=['label'],axis=1):
    np.savetxt('expPREs/exp{:d}.dat'.format(label),np.c_[df[label].value,df[label].error])
```

```python
def loadExpPREs(folder,labels,resnums):
    value = {}
    error = {}
    for label in labels:
        value[label], error[label] = np.loadtxt(folder+'/exp{:d}.dat'.format(label),unpack=True)
    v = pd.DataFrame(value,index=resnums,)
    v.rename_axis('residue', inplace=True)
    v.rename_axis('label', axis='columns',inplace=True)
    e = pd.DataFrame(error,index=resnums)
    e.rename_axis('residue', inplace=True)
    e.rename_axis('label', axis='columns',inplace=True)
    return pd.concat(dict(value=v,error=e),axis=1)
```

```python
fig, axes = plt.subplots(10, 1, sharex=True, sharey=True, figsize=(6,20))

exp = loadExpPREs('expPREs',labels,np.arange(1,221))
for ax,label in zip(axes.flatten(),exp.value):
    ax.errorbar(exp.index,exp.value[label],exp.error[label], color=plt.cm.tab10(0), lw=0, 
        marker='o', ms=0, elinewidth=2, capsize=4, capthick=1.5) 
    ax.set_xticks(range(0,exp.index.max()+1,30)[1:])
    ax.set_ylim(0,50)
    ax.annotate(str(label),xy=(40,40))
    ax.set_ylabel('$\Gamma_2$  /  s$^{-1}$')
    
axes.flatten()[-1].set_xlabel('Residue #')
```
